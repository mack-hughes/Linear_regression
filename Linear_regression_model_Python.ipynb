# Linear regression model in Python

Our goal is find out the linear model $$\hat y = \mathbf a \cdot \mathbf x + b$$ that best fit with the data

```julia
# Importing the libraries
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```julia
# Declaring the variables that controls the size of the array and the confidence interval.
number_of_points = 100
noise = 20
seed = 42
```

```julia
x, y = make_regression(
    n_samples=number_of_points, n_features=1, noise=noise, random_state=seed
)
plt.scatter(x, y)
plt.show()
```

```julia
# Criating the model
model = LinearRegression(fit_intercept=True, n_jobs=4)
```

```julia
# Passing the values x and y to the model
model.fit(x, y)
```

```julia
# Picking the coeficient a
a = model.coef_[0]
a
```

```julia
# Picking the coeficient b
b = model.intercept_
```

Now we can define our function

```julia
def y_hat(x):
    return a * x + b
```

```julia
x_hat = np.linspace(-3, 3, number_of_points)
plt.scatter(x, y)
plt.plot(x_hat, y_hat(x_hat), color="green")
plt.show()
```

We can analyse the $\color{red}coefficient \ of\ determination$ $r^2$. This term shows us how well this regression explains the data, it varies between 0 and 1, if $r^2$ = 0.84 it means that our linear model explains $84\%$ of the variation, it means that our linear model is $84\%$ more precise than just use the mean, the $\bar y$. 

$r^2 = 1 -\dfrac{SS_{res}}{SS_{tot}}$,

with the residual sum of squares being
$SS_{res} = \sum_i(\hat y ^i - y^i)^2$,

and the total sum of squares being
$SS_{tot} = \sum_i(\hat y ^i - \bar y)^2$,

where
 $\bar{y} = \dfrac{1}{n} \sum_{i=1}^n y^i$.

```julia
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```julia
model.fit(x_train, y_train)
```

```julia
model.score(x_test, y_test)
```

comparing $\hat y$ with $\bar y$

```julia
x_hat = np.linspace(-3, 3, number_of_points)
y_bar = 1 / len(y) * sum(y)
plt.scatter(x, y)
plt.plot(x_hat, y_hat(x_hat), color="green")
plt.plot(x_hat, np.full(len(y), y_bar), color="red")
plt.show()
```
